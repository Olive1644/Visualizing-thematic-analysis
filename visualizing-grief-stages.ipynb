```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import friedmanchisquare, wilcoxon
from statsmodels.stats.multitest import multipletests
import itertools

# Load data
file_path = "/kaggle/input/grief-count/Grief_Stage_Counts_Per_Participant.csv"
df = pd.read_csv(file_path)

# Define stages of interest
stages = ["Denial", "Anger", "Bargaining", "Depression", "Acceptance"]
data = df[stages]

# Friedman test (non-parametric repeated measures ANOVA)
friedman_result = friedmanchisquare(*[data[stage] for stage in stages])
print("Friedman test p-value:", friedman_result.pvalue)

# If significant, do pairwise Wilcoxon signed-rank tests
if friedman_result.pvalue < 0.05:
    p_values = []
    pairs = list(itertools.combinations(stages, 2))
    
    for a, b in pairs:
        stat, p = wilcoxon(data[a], data[b])
        p_values.append(p)

    # Holm-Bonferroni correction
    reject, p_corrected, _, _ = multipletests(p_values, method='holm')
    
    sig_pairs = [pair for pair, sig in zip(pairs, reject) if sig]
    print("Significant pairs after correction:", sig_pairs)
else:
    sig_pairs = []

# Plotting the means
means = data.mean()
colors = ['blue', 'red', 'orange', 'purple', 'green']

plt.figure(figsize=(10, 6))
bars = plt.bar(stages, means, color=colors)
plt.grid(False)
plt.title("MEAN GRIEF STAGES WITH SIGNIFICANT DIFFERENCES")
plt.ylabel("Mean Count")

# Annotate bar values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.3, f"{yval:.1f}", ha='center', va='bottom')

# Draw significance lines
y_max = max(means) + 1
h = 0.5  # height of the significance line
for i, (a, b) in enumerate(sig_pairs):
    x1, x2 = stages.index(a), stages.index(b)
    y = y_max + i * h
    plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], color='black')
    plt.text((x1 + x2)/2, y + h + 0.1, "*", ha='center')

plt.tight_layout()
plt.savefig("Grief_Stages_Significant_Differences.png", format='png', dpi=300)
plt.show()
```
